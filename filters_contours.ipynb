```python
import cv2
import matplotlib.image as mpimg
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
from google.colab.patches import cv2_imshow
from PIL import Image, ImageFilter

drive.mount('/content/drive', force_remount=True)
img=cv2.imread('/content/drive/MyDrive/T.jpg')
img=cv2.resize(img,(256,256))
plt.imshow(img[:,:,::-1])
plt.axis("off")
plt.title("Original")


```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
plt.axis("off")
plt.title("Grayscale")
```

```python
gray_inv=cv2.bitwise_not(img_gray)
plt.imshow(gray_inv,cmap='gray')
plt.axis("off")
plt.title("Inverted grayscale")
```

```python

contours,hierarchy=cv2.findContours(gray_inv,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
img_copy=img.copy()
cv2.drawContours(img_copy,contours,-1,(0,255,0),2)
plt.figure(figsize=[10,10])
plt.imshow(img_copy[:,:,::-1])
```

```python
binary=cv2.threshold(gray_inv,50,255,cv2.THRESH_BINARY)
plt.imshow(binary,cmap='gray')
```

```python
blurred=cv2.GaussianBlur(img,(5,5),0)

edges=cv2.Canny(blurred,100,200)
cv2_imshow(img)
print('\n')
cv2_imshow(edges)
```

```python
contours,hierarchy=cv2.findContours(edges,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
img_copy=img.copy()
cv2.drawContours(img_copy,contours[20],-1,(0,255,0),2)
cv2_imshow(img)
plt.figure(figsize=[3,3])
plt.imshow(img_copy[:,:,::-1])
plt.axis("off")
```
